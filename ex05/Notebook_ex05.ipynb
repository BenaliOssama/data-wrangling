```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    'value': [20.45, 22.89, 32.12, 111.22, 33.22, 100, 99.99],
    'product': ['table', 'chair', 'chair', 'mobile phone', 'table', 'mobile phone', 'table']
})

result = df.groupby('product').agg({'value': ['min', 'max', 'mean']})
print(result.to_markdown())
```
