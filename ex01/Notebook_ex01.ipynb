```python
import pandas as pd
```

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'number'])

df2 = pd.DataFrame([['c', 1], ['d', 2]],
                   columns=['letter', 'number'])

# Stack them, then reset the index
result = pd.concat([df1, df2]).reset_index(drop=True) # don't add the old index as a column, just discard it.

print(result)
print(result.index)
```
