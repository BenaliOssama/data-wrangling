```python
import pandas as pd
```

```python
df1_dict = {
    'id': ['1', '2', '3', '4', '5'],
    'Feature1': ['A', 'C', 'E', 'G', 'I'],
    'Feature2': ['B', 'D', 'F', 'H', 'J']}

df1 = pd.DataFrame(df1_dict, columns=['id', 'Feature1', 'Feature2'])

df2_dict = {
    'id': ['1', '2', '6', '7', '8'],
    'Feature1': ['K', 'M', 'O', 'Q', 'S'],
    'Feature2': ['L', 'N', 'P', 'R', 'T']}

df2 = pd.DataFrame(df2_dict, columns=['id', 'Feature1', 'Feature2'])
```

```python
# inner — only keep rows where the id exists in both tables → ids 1, 2
# outer — keep everything from both tables, fill gaps with NaN → ids 1,2,3,4,5,6,7,8
# left — keep everything from the left table, fill gaps from right with NaN
# right — the opposite
```

```python
# Question 1: inner merge, default suffixes _x and _y
q1 = pd.merge(df1, df2, on='id', how='inner') # By default it adds _x and _y
print(q1.to_markdown()) #  formats a DataFrame as a markdown table
```

```python
# Question 2: outer merge, custom suffixes
q2 = pd.merge(df1, df2, on='id', how='outer', suffixes=('_df1', '_df2'))
print(q2.to_markdown()) #  formats a DataFrame as a markdown table
```
