```python
import pandas as pd
import numpy as np
```

```python
def winsorize(df, quantiles):
    min_value = np.quantile(df, quantiles[0]) #what value sits at this percentile in my data
    max_value = np.quantile(df, quantiles[1])
    return df.clip(lower=min_value, upper=max_value) #  enforce a floor and a ceiling on every value.
```

```python
df = pd.DataFrame(range(1,11), columns=['sequence'])
print(winsorize(df, [0.20, 0.80]).to_markdown())
```

```python
groups = np.concatenate([np.ones(10), np.ones(10)+1,  np.ones(10)+2, np.ones(10)+3, np.ones(10)+4])

df = pd.DataFrame(data= zip(groups,
                            range(1,51)),
                columns=["group", "sequence"])
df.head()
```

```python
#go 5% of the way from the bottom" and "go 95% of the way from the top.
result = df.groupby("group")[['sequence']].apply(winsorize, [0.05, 0.95])
print(result.head(11).to_markdown())
```
