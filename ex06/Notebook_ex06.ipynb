```python
import pandas as pd
import numpy as np
```

```python
business_dates = pd.bdate_range('2021-01-01', '2021-12-31')

# generate tickers
tickers = ['AAPL', 'FB', 'GE', 'AMZN', 'DAI']

# create indexes
index = pd.MultiIndex.from_product([business_dates, tickers], names=['Date', 'Ticker'])

# create DFs
market_data = pd.DataFrame(index=index,
                        data=np.random.randn(len(index), 1),
                        columns=['Prediction'])
market_data
```

```python
# Question 1: unstack the Ticker level into columns
unstacked = market_data.unstack()
print(unstacked.head(3).to_markdown())
```

```python
# Question 2: plot all 5 time series
unstacked.plot(title='Stocks 2021')
```
